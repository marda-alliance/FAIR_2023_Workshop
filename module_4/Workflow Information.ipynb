## Access Information and Other Configurations

```python
import base64
import json
import requests


access_payload = input("Please input your HyperThought API Access Payload: ")
access_object = json.loads(base64.b64decode(access_payload))

headers = {
    'Authorization': f'Bearer {access_object["accessToken"]}'
}
```

## Get Workflow Information from HyperThought

```python
workflow_pk = input("Workflow Identifier: ")

workflow_info_request = requests.get(
    url=f'{access_object["baseUrl"]}/api/workflow/process/{workflow_pk}/',
    headers=headers,
    verify=False
)

workflow_info_request_data = workflow_info_request.json()

print(json.dumps(workflow_info_request_data, indent=2))
```

## Get Workflow Children

```python
workflow_info_children_request = requests.get(
    url=f'{access_object["baseUrl"]}/api/workflow/process/{workflow_pk}/children',
    headers=headers,
    verify=False
)

workflow_info_children_request_data = workflow_info_children_request.json()

print(json.dumps(workflow_info_children_request_data, indent=2))
```
